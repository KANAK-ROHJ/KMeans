```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
X,y=make_blobs(n_samples=500,n_features=2,centers=5,random_state=3)
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
plt.grid(True)
plt.show()
```

### Random initialisation of centers

```python
k=5
colors=['green','yellow','blue','cyan','red']
clusters={}
for i in range(k):
    center=10*(2*np.random.random((2,))-1)
    points=[]
    cluster={
        'center':center,
        'points':points,
        'color' :colors[i],
    }
    clusters[i]=cluster
```

```python
clusters
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
for i in clusters:
    center=clusters[i]['center']
    plt.scatter(center[0],center[1],marker='^',c='red')
plt.show()
```

```python
def distance(p1,p2):
    return np.sqrt(np.sum((p1-p2)**2))
```

### Assign each point to nearest center  (E-step)

```python
def assign_clusters():
    for idx in range(X.shape[0]):
        dist=[]
        p=X[idx]
        for i in range(k):
            d=distance(p,clusters[i]['center'])
            dist.append(d)
        cluster=np.argmin(dist)
        clusters[cluster]['points'].append(p)
```

### Update center location by taking mean

```python
def update_clusters():
    for idx in range(k):
        pts=np.array(clusters[idx]['points'])
        if pts.shape[0]>0:
            new_center=pts.mean(axis=0)
            clusters[idx]['center']=new_center
            clusters[idx]['points']=[]
            
```

```python
def plot_clusters():
    for i in clusters:
        pts=np.array(clusters[i]['points'])
        try: 
            plt.scatter(pts[:,0],pts[:,1],c=clusters[i]['color'])
        except:
            pass
        center=clusters[i]['center']
        plt.scatter(center[0],center[1],c='black',marker='^')
```

```python
assign_clusters()
plot_clusters()
update_clusters()
```

# Scikit learn K-Means

```python
from sklearn.cluster import KMeans
```

```python
km=KMeans(n_clusters=5)
```

```python
km.fit(X,y)
```

```python
centers=km.cluster_centers_
```

```python
labels=km.labels_
```

```python
plt.scatter(X[:,0],X[:,1],c=labels)
plt.scatter(centers[:,0],centers[:,1],c='black',marker="^")
plt.show()
```
